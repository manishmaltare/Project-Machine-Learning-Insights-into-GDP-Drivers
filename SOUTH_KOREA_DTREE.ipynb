```python
x = df.drop(['GDP'],axis=1)
```

```python
x.head(2)
```

```python
y=df['GDP']
```

```python
# Applying the ML algorithms
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42, train_size=0.8)
```

# DECISION TREE

### Training Dataset

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtree_train= DecisionTreeRegressor()
```

```python
model_dtree_train = dtree_train.fit(x_train,y_train)
```

```python
y_pred_train = model_dtree_train.predict(x_train)
```

### Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
r2_score_train=r2_score(y_pred_train, y_train)

print(f"R2-score : {r2_score_train}")
```

### Testing Dataset

```python
dtree_test= DecisionTreeRegressor()
```

```python
model_dtree_test = dtree_test.fit(x_test,y_test)
```

```python
y_pred_test = model_dtree_test.predict(x_test)
```

### Evaluation

```python
r2_score_test=r2_score(y_pred_test, y_test)

print(f"R2-score : {r2_score_test}")
```
